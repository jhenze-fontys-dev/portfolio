## 1) API key (gegeven)

```python
# Exact gegeven key (niet aanpassen):
# VITE_TWELVE_DATA_KEY=f49a99207d6f4678922a730cdac72bf0

import os, requests, json
import pandas as pd

os.environ["VITE_TWELVE_DATA_KEY"] = "f49a99207d6f4678922a730cdac72bf0"
API_KEY = os.environ["VITE_TWELVE_DATA_KEY"]
BASE_URL = "https://api.twelvedata.com"

def show_json(data):
    print(json.dumps(data, indent=2)[:4000])  # max ~4000 tekens voor overzicht

def td_get(endpoint, **params):
    params["apikey"] = API_KEY
    r = requests.get(f"{BASE_URL}/{endpoint}", params=params, timeout=30)
    r.raise_for_status()
    return r.json()

print("Ready ✅")
```

## 2) Kies een symbool

Kies één van deze (of je eigen):
- `AAPL`, `MSFT`, `TSLA`, `NFLX`, `AMZN`
- `ASML` (werkt soms met exchange erbij, afhankelijk van de data)

Vul hieronder het symbool in.


```python
Symbol = "NFLX"  # <-- leerling vult hier in
Symbol
```

## 3) Quote (huidige info / snelle snapshot)

```python
data = td_get("quote", symbol=Symbol)
show_json(data)
```

## 4) Time series (OHLC) – 1 dag interval

```python
data = td_get("time_series", symbol=Symbol, interval="1day", outputsize=30)
show_json(data)
```

## 5) Time series omzetten naar een tabel (pandas)

Dit maakt het makkelijker om kolommen te zien (`open`, `high`, `low`, `close`, `volume`, ...).


```python
data = td_get("time_series", symbol=Symbol, interval="1day", outputsize=30)

df = pd.DataFrame(data.get("values", []))
df.head()
```

## 6) Andere calls om uit te proberen (run er één tegelijk)

Probeer hieronder verschillende endpoints:

- `price` (alleen prijs)
- `exchanges` (lijst met exchanges)
- `symbol_search` (zoek symbolen op naam)
- `earnings` (als beschikbaar voor het symbool)


```python
# A) Alleen de prijs
data = td_get("price", symbol=Symbol)
show_json(data)
```

```python
# B) Exchanges (kan veel output geven)
data = td_get("exchanges")
show_json(data)
```

```python
# C) Symbol search (zoek op tekst)
zoekterm = "netflix"   # <-- leerling past aan
data = td_get("symbol_search", symbol=zoekterm)
show_json(data)
```

```python
# D) Earnings (niet voor elk symbool beschikbaar)
data = td_get("earnings", symbol=Symbol)
show_json(data)
```
